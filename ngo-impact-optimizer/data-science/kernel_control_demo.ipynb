# Kernel Control Demonstration

This notebook demonstrates controlled execution of cells, interrupting a running cell, and restarting the kernel to reset the execution state.

## Step 1: Sequential Execution

In this step, we execute multiple code cells in order to demonstrate how the kernel maintains state between executions.

```python
a = 10
print("Value of a:", a)
```

```python
b = a + 5
print("Value of b:", b)
```

## Step 2: Long Running Cell (Interrupt Demonstration)

This cell runs continuously and will be manually interrupted to demonstrate kernel control.

```python
import time

while True:
    print("Running...")
    time.sleep(1)
```

## Step 3: Kernel Restart

After interrupting execution, the kernel will be restarted to clear all variables and reset the environment.

## Step 4: Re-execution After Restart

After restarting the kernel, all previous variables are cleared. Cells must be executed again from the beginning to restore state.

```python
print(a)
```

```python
a = 10
b = a + 5
print("Recovered values:", a, b)
```
